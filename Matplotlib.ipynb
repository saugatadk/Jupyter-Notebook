# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib auto
```

```python
import numpy as np
```

```python
x = np.arange(1, 11)
y = np.arange(11, 21)
a = np.arange(50, 60)
b = np.arange(60, 70)
```

```python
x, y, a, b
```

```python
plt.scatter(x, y, c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Plot in 2D")
plt.grid(True)
```

```python
y=x**3
```

```python
plt.plot(x, y, 'r*', linestyle='dashed')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
plt.grid(True)
```

```python
plt.subplot(2,2,1)
plt.plot(x, y, 'r')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.subplot(2,2,4)
plt.plot(x,y,'yo-')
```

```python
np.pi
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.grid(True)
plt.title("Sine Wave")
```

```python
x1 = [1,3,5]
y1 = [5,6,7]

x2 = [2,4,6]
y2 = [20, 2, 14]
plt.bar(x1, y1)
plt.bar(x2, y2, color='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Bar Graph")
```

# Histograms

```python
a = np.array([22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27])
plt.hist(a)
plt.grid(True)
plt.title("Histogram")
```

```python
data = [1, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5, 5, 6, 7, 8]

# plt.hist(data, bins=4)  # 4 bins  data grouped into 4 intervals
plt.hist(data, bins=[1, 3, 5, 7, 9])
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct= "%1.1f%%", shadow=True)
plt.axis('equal')
```
