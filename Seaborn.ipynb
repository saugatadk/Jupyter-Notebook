# Seaborn Library

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
corr_data = df.select_dtypes(include='number').corr()
corr_data
```

```python
sns.heatmap(corr_data)
```

# JoinPlot
A join plot allows to study the relationship between 2 numbers. The central chart represents correlation. It is usually a scatterplot, a hexbin plot , a 2D histogram or a 2D density plot.

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

# PairPlot
Used if more than 2 independent features are present. It is like a scatterplot and does all the permutation and combination of all the features

```python
sns.pairplot(df)
```

```python
df['sex'].value_counts()
```

```python
sns.pairplot(df, hue='sex')
```

# Dist Plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10) #This creates a smooth curve instead of bars, helping to see patterns and peaks more clearly than a histogram.
```

# Categorical Plots
Seaborn also helps us in the analysis on Categorical Data Points
1. boxplot
2. violinplot
3. countplot
4. bar plot

```python
sns.countplot(x='sex', data=df )
```

```python
sns.countplot(x='day', data=df)
```

```python
sns.countplot(y='day', data=df)
```

```python
sns.barplot(x='day', y='total_bill', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df )
```

# Box Plot
A box and whisker plot(sometimes called a boxplot) is a graph that 
represents information from 5 number summary

```python
sns.boxplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='sex', y='total_bill', hue='sex', legend=False, data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='v') #on complete dataset
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df) #categorize my data based on some other categories
```

# Violin Plot
Violin plot helps us to see both the distribution of data in terms of 
Kernel Density Estimation(KDE) and the box plot

```python
sns.violinplot(x='total_bill', y='day', hue='day', data=df, palette='rainbow')
```
